```python
class Car: ##def is used for defining function and init stands for initializing and 2 underscroll is necessary
    def __init__(self, model, year, color, for_sale):
        self.model = model
        self.year = year
        self.color = color
        self.for_sale = for_sale


car1 = Car("Tesla", 2022, "Red", True)
car2 = Car("BMW", 2023, "Black", False)
car3 = Car("Audi", 2024, "Blue", True)


print(car3.model)
print(car3.year)
print(car3.color)
print(car3.for_sale)



```
